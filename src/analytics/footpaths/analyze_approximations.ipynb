```python
def toId(uri):
    id = uri.split("//")[1];
    id = id.replace(".", "_");
    return id.replace("/", "_");
```

```python
def to_file_name(t, f):
    dir_name = "distances/{}/".format(toId(f))
    return "{}{}.json".format(dir_name, toId(t)) 
```

```python
def load_distances(pairs):
    distances = []
    for f, t in pairs:
        try:
            with open(to_file_name(f, t)) as file:
                distance = int(file.read())
                distances.append((f, t, distance))
        except:
            pass
        
        try:
            with open(to_file_name(t, f)) as file:
                distance = int(file.read())
                distances.append((t, f, distance))
        except:
            pass
    return distances
```

```python
nmbs_distances = load_distances(nmbs_pairs)
```

```python
tec_distances = load_distances(tec_pairs)
```

```python
delijn_distances = load_distances(delijn_pairs)
```

```python
mivb_distances = load_distances(mivb_pairs)
```

```python
combined_distances = load_distances(combined_pairs)
```

```python
from collections import defaultdict
from heapq import *

def dijkstra(edges, f, t):
    g = defaultdict(list)
    for l,r,c in edges:
        g[l].append((c,r))

    q, seen, mins = [(0,f)], set(), {f: 0}
    while q:
        (cost,v1) = heappop(q)
        if v1 not in seen:
            seen.add(v1)
            if v1 == t: 
                return cost

            for c, v2 in g.get(v1, ()):
                if v2 in seen: 
                    continue
                prev = mins.get(v2, None)
                next = cost + c
                if prev is None or next < prev:
                    mins[v2] = next
                    heappush(q, (next, v2))
    return float("inf")
```

```python
from haversine import haversine
import requests

def load_stations():
    result = {}
    stations = json.loads(requests.get("https://irail.be/stations/NMBS").content)["@graph"]
    for station in stations:
        result[station["@id"]] = float(station['http://www.w3.org/2003/01/geo/wgs84_pos#lat']), \
            float(station['http://www.w3.org/2003/01/geo/wgs84_pos#long'])
    return result

stations = load_stations()
```

```python
from tqdm import tqdm

triangle_errors = []
haversine_errors = []

edges = delijn_distances + mivb_distances + tec_distances + combined_distances
for f, t, actual in tqdm(nmbs_distances):
    if f < t:
        approximated = dijkstra(edges, f, t)
        if approximated != float("inf"):
            triangle_errors.append((actual, approximated))
            
            approximated = haversine(stations[f], stations[t], 'm')
            haversine_errors.append((actual, approximated))
```

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
import pandas as pd
```

```python
actual_distance_bins = []
actual_distances = []
triangle_approximations = []
haversine_approximations = []

for h_error, t_error in zip(haversine_errors, triangle_errors):
    actual, h_predicted = h_error
    _, t_predicted = t_error
    
    h_relative_error = ((h_predicted / actual) - 1) * 100
    t_relative_error = ((t_predicted / actual) - 1) * 100
    if t_relative_error < 225:
        actual_distance_bins.append((actual + 500) // 1000)
        actual_distances.append(actual / 1000)
        triangle_approximations.append(t_relative_error)
        haversine_approximations.append(h_relative_error)
```

```python
combined_approximations = [(x + y) / 2 for x, y in zip(triangle_approximations, haversine_approximations)]
```

```python
x, y = pd.Series(actual_distance_bins, name="Actual Distance (km)"), pd.Series(triangle_approximations, name="Over-estimation (%)")
ax = sns.regplot(x=x, y=y, color='black', x_estimator=np.mean)

plt.title('Distance approximation error')
plt.tight_layout()
plt.savefig('overestimation.svg')
```
